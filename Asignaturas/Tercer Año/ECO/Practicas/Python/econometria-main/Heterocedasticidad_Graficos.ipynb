```python

import pandas as pd #librería para manejo de datos

datos= pd.read_csv("https://rtgodwin.com/data/houseprice.csv") #Lee base de datos de web...
```

```python
import statsmodels.api as sm

y=datos["Price"]
X=datos[['Lot.Size', 'Age', 'Land.Value',   'Living.Area', 'Bedrooms']]

mco = sm.OLS(y, sm.add_constant(X)).fit()
print(mco.summary())
```

# Gráficos

```python
import matplotlib.pylab as plt

plt.scatter(list(range(int(mco.nobs))), mco.resid)
plt.title("Observaciones vs Residuos")
plt.show()

```

```python
plt.scatter(list(range(int(mco.nobs))), mco.resid**2)
plt.title("Observaciones vs Residuos^2")
plt.show()
```

```python
plt.scatter(datos["Lot.Size"], mco.resid)
plt.title("Lot.Size vs Res")
plt.show()
plt.scatter(datos["Age"], mco.resid)
plt.title("Age vs Res")
plt.show()
plt.scatter(datos["Land.Value"], mco.resid)
plt.title("Land.Value vs Res")
plt.show()
plt.scatter(datos["Living.Area"], mco.resid)
plt.title("Living.Area vs Res")
plt.show()
plt.scatter(datos["Bedrooms"], mco.resid)
plt.title("Bedrooms vs Res")
plt.show()
```
