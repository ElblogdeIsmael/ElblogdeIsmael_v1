```python
import pandas as pd #librerÃ­a para manejo de datos

datos= pd.read_csv("https://rtgodwin.com/data/houseprice.csv") #Lee base de datos de web...

datos.columns
```

```python
import statsmodels.api as sm


y = datos["Price"]
X = datos[['Lot.Size', 'Waterfront', 'Age', 'Land.Value', 'New.Construct',
       'Central.Air', 'Living.Area',
       'Pct.College', 'Bedrooms', 'Fireplaces', 'Bathrooms', 'Rooms']]

mco = sm.OLS(y, sm.add_constant(X)).fit()

print(mco.summary())
```

```python
import numpy as np

datos["logPrecio"] = np.log(datos["Price"])
datos["logLandValue"] = np.log(datos["Land.Value"])
```

```python
y = datos["logPrecio"]
X = datos[['Lot.Size', 'Waterfront', 'Age', 'Land.Value', 'New.Construct',
       'Central.Air', 'Living.Area',
       'Pct.College', 'Bedrooms', 'Fireplaces', 'Bathrooms', 'Rooms']]

mco = sm.OLS(y, sm.add_constant(X)).fit()

print(mco.summary())
```
