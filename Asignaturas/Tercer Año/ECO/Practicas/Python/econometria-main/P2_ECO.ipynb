```python
import pandas as pd


datos = pd.read_excel('/content/drive/MyDrive/AQUI SE PONE EL NOMBRE DE LA CARPETA DEL GOOGLE DRIVE/Y AQUI SE PONE EL NOMBRE DEL FICHEROCOMOSELLAME')
```

```python
datos
```

```python
dumis_nuevas= pd.get_dummies(datos["MYVARIABLE"])
datos = pd.concat([datos, dumis_nuevas], axis=1)
datos
```

```python
import statsmodels.api as sm

X=datos[['artist_count', 'released_year', 'released_month',
       'released_day', 'in_spotify_playlists', 'in_spotify_charts',
       'in_apple_playlists', 'in_apple_charts', 'in_deezer_charts', 'bpm',
       'danceability_%', 'valence_%', 'energy_%', 'acousticness_%',
       'instrumentalness_%', 'liveness_%', 'speechiness_%']]
y=datos["streams"]/100000


results = sm.OLS(y, sm.add_constant(X)).fit()

print(results.summary())
```
